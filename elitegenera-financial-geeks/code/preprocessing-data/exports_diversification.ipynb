```python
df = df.drop(0)
df = df.reset_index(drop=True)
df.head()
```

```python
max_export_products = []
for year in df.columns[2:]:
    max_export_product = df.loc[df[year].idxmax()]['Product label']
    max_export_products.append((year, max_export_product))

# Displaying the results
for year, product in max_export_products:
    print(f"Highest {year}: {product}")
```

```python
min_export_products = []
for year in df.columns[2:]:
    min_export_product = df.loc[df[year].idxmin()]['Product label']
    min_export_products.append((year, min_export_product))

# Displaying the results
for year, product in min_export_products:
    print(f"Loweest {year}: {product}")
```

```python
missing_values = df.isnull().sum()

print("Missing values count per column:")
print(missing_values)
```

```python
# Calculate percentage change for each year
df_percentage_change = df.pct_change(axis=1) * 100

# Display the result
print("Percentage Change in Exports from Year to Year:")
print(df_percentage_change)
```

```python
#Diversification index
# Calculate the market share for each product in each year
market_share = df.div(df.sum(axis=0), axis=1)

# Calculate the HHI for each year,The Herfindahl-Hirschman Index (HHI) is commonly used to measure the concentration or diversification of exports. The formula for the HHI is the sum of the squared market shares of each product.
hhi = (market_share ** 2).sum()

# Display the result
print("Herfindahl-Hirschman Index (HHI) for each year:")
print(hhi)
```

```python
# Set 'Code' and 'Product label' as index
df.set_index(['Code', 'Product label'], inplace=True)

# Calculate yearly growth rates for each product
growth_rates = df.pct_change(axis=1) * 100

# Define a threshold for substantial growth (e.g., 10%)
growth_threshold = 10

# Identify emerging products with substantial growth in recent years
emerging_products = growth_rates[growth_rates.abs() > growth_threshold].dropna(how='all')

# Display the result
print("Emerging Products with Substantial Growth:")
print(emerging_products)
```
